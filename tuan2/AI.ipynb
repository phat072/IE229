# 問題3

```python
import numpy as np
import matplotlib.pyplot as plt
# Data D
X=np.array([0.349526784,1.6974435,5.384308891,2.044150596,
4.578814506,3.241690807,2.535931731,2.210580888,
3.397474351,5.972933146,5.114704101])
Y=np.array([0.254020646,0.790556868,-0.81239532,1.012143475,
-0.904558188,-0.167456361,0.482547054,0.878514378,
-0.210093715,-0.128786937,-0.866501299])

#a. least_square method 
X3 = []
for i in X:
    X3.append([1, i, i**2, i**3])

X3 = np.array(X3)

z1 = np.matmul(X3.T, X3)
z2 = np.linalg.inv(z1)
z3 = np.matmul(z2, X3.T)

w = np.matmul(z3, Y)
plt.scatter(X, Y, color="red")

def FSin(x):
    return np.sin(x)

x_min = 0
x_max = 6
x_values = np.linspace(x_min, x_max, 100)
y_values = FSin(x_values)
plt.plot(x_values, y_values, color="green")

def FCubic(x):
    return w[3]*x**3 + w[2]*x**2 + w[1]*x + w[0]

x_new = np.linspace(x_min, x_max, 50)
y_new = FCubic(x_new)
plt.plot(x_new, y_new, color="blue")
plt.show()
```

```python
#b. 9th-degree polynomial
X3 = []
for i in X:
    X3.append([1, i, i**2, i**3, i**4, i**5, i**6, i**7, i**8, i**9])
    
X3 = np.array(X3)
Y3 = np.array(Y)

z1 = np.matmul(X3.T, X3)
z2 = np.linalg.inv(z1)
z3 = np.matmul(z2, X3.T) 

w = np.matmul(z3, Y3)

plt.scatter(X, Y, color="red")
def FSin(x):
    return np.sin(x)

x_min = 0
x_max = 6
x_values = np.linspace(x_min, x_max, 100)
y_values = FSin(x_values)
plt.plot(x_values, y_values, color="green")

def FCubic(x):
    return w[9]*x**9 + w[8]*x**8 + w[7]*x**7 + w[6]*x**6 + w[5]*x**5 + w[4]*x**4 + w[3]*x**3 + w[2]*x**2 + w[1]*x + w[0]

x_new = np.linspace(x_min, x_max, 500)
y_new = FCubic(x_new)
plt.plot(x_new, y_new, color="blue")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data D
X=np.array([0.349526784,1.6974435,5.384308891,2.044150596,
4.578814506,3.241690807,2.535931731,2.210580888,
3.397474351,5.972933146,5.114704101])

Y=np.array([0.254020646,0.790556868,-0.81239532,1.012143475,
-0.904558188,-0.167456361,0.482547054,0.878514378,
-0.210093715,-0.128786937,-0.866501299])

# Generate x values
x_values = np.linspace(0, 6, 100)

# Compute sine of x values
y_values = np.sin(x_values)

# Plot sine function
plt.plot(x_values, y_values, color="green")

# Scatter plot of data D
plt.scatter(X, Y, color="red")

# Fit 9th-degree equation and plot
coeffs_9 = np.polyfit(X, Y, 9)
poly_9 = np.poly1d(coeffs_9)
plt.plot(x_values, poly_9(x_values), color="blue")
plt.show()
```
