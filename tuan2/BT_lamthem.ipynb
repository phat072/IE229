### Câu 2. Vẽ biểu đồ đường biểu diễn tổng lợi nhuận theo từng tháng, theo định dạng sau:
+   Nét vẽ màu đỏ, độ dày là 3 và là đường đứt nét.
+   Trục X biểu diễn các tháng.
+   Trục Y biểu diễn tổng lợi nhuận của mỗi tháng.
+   Các điểm dữ liệu được vẽ bằng hình tròn và có màu đen.

```python
import matplotlib.pyplot as plt

# Calculate total profit by month
total_profit = df['total_profit'].values
plt.figure(figsize=(10, 6))

# Create an array of months from 1 to 12
months = range(1, 13)

# Draw the line chart
plt.plot(months, total_profit, color='red', linewidth=3, linestyle='dashed', marker='o', markerfacecolor='black')

# Set the name for the x-axis and y-axis
plt.xlabel('Month')
plt.xticks(months)
plt.ylabel('Total Profit')
plt.title("Biểu đồ đường biểu diễn tổng lợi nhuận theo từng tháng")

# Display the chart
plt.show()
```

### Câu 3. Vẽ biểu đồ đường biểu diễn số lượng đã bán của mỗi sản phẩm theo từng tháng, theo định
dạng sau:
+   Đường biểu diễn của các sản phẩm khác màu nhau.
+   Nét vẽ có chiều rộng là 3 và là đường liền nét.
+   Bảng chú thích các đại lượng nằm ở vị trí phía trên bên trái.
+   Trục X biểu diễn các tháng.
+   Trục Y biểu diễn số lượng sản phẩm đã bán.
+   Các điểm dữ liệu được vẽ bằng hình tròn và có cùng màu với đường biểu diễn

```python
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(12, 8))

product_names = df.columns[1:7] # The first column is 'month'

# Create an array of months from 1 to 12
months = range(1, 13)

# For each product, draw a line chart
for product in product_names:
    monthly_sales = df[product].values
    plt.plot(months, monthly_sales, linewidth=3, marker='o')

# Set the name for the x-axis and y-axis
plt.xlabel('Month')
plt.xticks(months)
plt.ylabel('Quantity Sold')

# Set the legend in the upper left corner
plt.legend(product_names, loc='upper left')
plt.title("Biểu đồ đường biểu diễn số lượng đã bán của mỗi sản phẩm theo từng tháng")

# Display the chart
plt.show()
```

### Câu 4. Vẽ biểu đồ phân tán biểu diễn số lượng sản phẩm kem đánh răng đã bán theo từng tháng,
theo định dạng sau:

+   Hiển thị đường lưới có độ dày là 1 và là nét đứt

```python
import matplotlib.pyplot as plt

# Get the quantity of toothpaste sold each month
toothpaste_sales = df['toothpaste'].values
plt.figure(figsize=(12, 8))

# Create an array of months from 1 to 12
months = range(1, 13)

# Create a scatter plot
plt.scatter(months, toothpaste_sales)

# Set the name for the x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Quantity Sold')
plt.xticks(months)

# Display the grid with dashed line style
plt.grid(linestyle='dashed')

plt.title("Biểu đồ phân tán biểu diễn số lượng sản phẩm kem đánh răng đã bán theo từng tháng")
# Display the chart
plt.show()
```

Câu 5. Vẽ biểu đồ cột đôi biểu diễn số lượng sản phẩm kem dưỡng da và sữa rửa mặt đã bán theo
từng tháng, theo định dạng sau:
+   Bảng chú thích các đại lượng nằm ở vị trí phía trên bên trái.
+   Trục X biểu diễn các tháng.
+   Trục Y biểu diễn số lượng sản phẩm đã bán.
+   Cột biểu diễn của hai loại sản phẩm khác màu nhau.
+   Hiển thị đường lưới có độ dày là 1 và là nét đứt.

```python
import matplotlib.pyplot as plt
import numpy as np

face_cream_sales = df['facecream'].values
face_wash_sales = df['facewash'].values
plt.figure(figsize=(12, 8))

months = range(1, 13)


bar_width = 0.35
bar_positions = np.arange(len(months))

# Create a bar chart for face cream sales
plt.bar(bar_positions, face_cream_sales, bar_width, label='Face Cream sales')
plt.bar(bar_positions + bar_width, face_wash_sales, bar_width, label='Face Wash sales')

# Set the name for the x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Quantity Sold')

plt.xticks(bar_positions + bar_width / 2, months)

# Set the legend in the upper left corner
plt.legend(loc='upper left')

# Display the grid with dashed line style
plt.grid(linestyle='dashed')

plt.title("Biểu đồ cột đôi biểu diễn số lượng sản phẩm kem dưỡng da và sữa rửa mặt đã bán theo từng tháng")
# Display the chart
plt.show()
```

### Câu 8. Vẽ biểu đồ tròn biểu diễn tỉ lệ phần trăm doanh thu của mỗi sản phẩm trong tổng thể
doanh thu của công ty, theo định dạng sau:
+   Mỗi sản phẩm biểu diễn bằng một màu khác nhau.
+   Tên mỗi sản phẩm hiển thị kèm theo phần biểu diễn tương ứng trên biểu đồ.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
# Calculate the total revenue for each product
product_names = df.columns[1:7]  # The first column is 'month'
total_revenue_per_product = [df[product].sum() for product in product_names]

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(total_revenue_per_product, labels=product_names, autopct='%1.1f%%')

# Display the chart
plt.title("Biểu đồ tròn biểu diễn tỉ lệ phần trăm doanh thu của mỗi sản phẩm trong tổng thể doanh thu của công ty")
plt.show()
```

### Câu 9. Vẽ biểu đồ gồm các biểu đồ con là các biểu đồ đường biểu diễn số lượng sản phẩm xà
phòng tắm và sữa rửa mặt đã bán theo từng tháng, theo định dạng sau:
+   Mỗi biểu đồ biểu diễn cho một sản phẩm, trong đó biểu đồ biểu diễn xà phòng tắm
nằm phía trên và biểu đồ biểu diễn sữa rửa mặt nằm phía dưới.
+   Đường biểu diễn của hai loại sản phẩm khác màu nhau.
+   Trục X biểu diễn các tháng.
+   Trục Y biểu diễn số lượng sản phẩm đã bán.

```python
import matplotlib.pyplot as plt

# Get the quantity of bathing soap and face wash sold each month
bathing_soap_sales = df['bathingsoap'].values
face_wash_sales = df['facewash'].values
plt.figure(figsize=(12, 8))
# Create an array of months from 1 to 12
months = range(1, 13)

# Create a figure and a set of subplots
fig, axs = plt.subplots(2)

# Create a line chart for bathing soap sales
axs[0].plot(months, bathing_soap_sales, color='blue')
axs[0].set_title('Bathing Soap Sales')
axs[0].set_xticks(months)
axs[0].set_ylabel('Quantity Sold')

# Create a line chart for face wash sales
axs[1].plot(months, face_wash_sales, color='orange')
axs[1].set_title('Face Wash Sales')
axs[1].set_xticks(months)
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Quantity Sold')

# Display the chart
plt.tight_layout()
plt.show()
```

### Câu 10. Vẽ biểu đồ miền xếp biểu diễn số lượng đã bán của mỗi sản phẩm theo từng tháng, theo
định dạng sau:
+   Miền biểu diễn của các sản phẩm khác màu nhau.
+   Bảng chú thích các đại lượng nằm ở vị trí phía trên bên trái.
+   Trục X biểu diễn các tháng.
+   Trục Y biểu diễn số lượng sản phẩm đã bán.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
product_names = df.columns[1:7]  # The first column is 'month'
product_sales = [df[product].values for product in product_names]

# Create an array of months from 1 to 12
months = range(1, 13)

# Create a figure
plt.figure(figsize=(10, 6))

# Create a stacked area chart
plt.stackplot(months, product_sales, labels=product_names)

# Set the name for the x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Quantity Sold')

plt.xticks(months)

# Set the legend in the upper left corner
plt.legend(loc='upper left')

# Display the chart
plt.title("Biểu đồ miền xếp biểu diễn số lượng đã bán của mỗi sản phẩm theo từng tháng")
plt.show()
```
